```python
# Visualizing the Numerical Columns (Hidden Input) and treating outliers

fig, ax = plt.subplots(ncols=2, nrows=14, figsize=(16, 40))

i = 0 #counter
for col in num_cols[1:]:

    # Removing Outliers
    Q3 = df[col].quantile(0.99)
    df = df[df[col] <= Q3]
    Q1 = df[col].quantile(0.01)
    df = df[df[col] >= Q1]

    sns.kdeplot(x=col, data=df, fill=True, alpha=1, hue = 'Attrition_Flag',
                palette='pastel', multiple='stack', ax=ax[i,0])

    sns.boxplot(data= df, y=col, x='Attrition_Flag', ax=ax[i, 1],
               palette='pastel', color='white')
    ax[i,0].set_xlabel(' ')
    ax[i,1].set_xlabel(' ')
    ax[i,1].set_ylabel(' ')
    ax[i,1].xaxis.set_tick_params(labelsize=14)
    ax[i,0].tick_params(left=False, labelleft=False)
    ax[i,0].set_ylabel(col, fontsize=12)
    i=i+1

plt.show()
```

This Feature is Imbalanced, hence, we need to fix this.
This can mislead many models into predicting “no churn” all the time, and still getting >80% accuracy.

```python
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
#ONE HOT ENCODING
```

```python
X = df.drop('Attrition_Flag', axis=1)
y = df['Attrition_Flag']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42, sampling_strategy=1.0)
X_train, y_train = sm.fit_resample(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        n_jobs=-1,
        random_state=42
)

rf.fit(X_train, y_train)
y_pred  = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

print(classification_report(y_test, y_pred, digits=3))
print("ROC-AUC:", roc_auc_score(y_test, y_proba).round(3))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```

The model is catching most churners (recall = 85.4%)

```python
pd.Series(rf.feature_importances_, index=X_train.columns)\
  .sort_values(ascending=False).head(15)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix

# Define the model
xgb = XGBClassifier(
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42,
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8
)

# 5-fold stratified cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(xgb, X_train, y_train, cv=cv, scoring='roc_auc')

print("Cross-validated ROC-AUC scores:", scores)
print("Average ROC-AUC:", scores.mean().round(4))

# Fit on full training data
xgb.fit(X_train, y_train)

# Evaluate on test set
y_pred = xgb.predict(X_test)
y_proba = xgb.predict_proba(X_test)[:, 1]

print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))
print("ROC-AUC:", roc_auc_score(y_test, y_proba).round(3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

param_dist = {
    'n_estimators': [100, 300, 500],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 1, 5],
    'reg_alpha': [0, 0.5, 1],
    'reg_lambda': [1, 1.5, 2]
}

xgb_base = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

random_search = RandomizedSearchCV(
    estimator=xgb_base,
    param_distributions=param_dist,
    n_iter=30,
    scoring='roc_auc',
    n_jobs=-1,
    cv=5,
    verbose=1,
    random_state=42
)

random_search.fit(X_train, y_train)

print("Best Parameters:", random_search.best_params_)
print("Best ROC-AUC from CV:", random_search.best_score_.round(4))

# Final model with best parameters
best_xgb = random_search.best_estimator_
y_pred_tuned = best_xgb.predict(X_test)
y_proba_tuned = best_xgb.predict_proba(X_test)[:, 1]

print("\nFinal Classification Report:\n", classification_report(y_test, y_pred_tuned, digits=3))
print("Final ROC-AUC:", roc_auc_score(y_test, y_proba_tuned).round(3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tuned))
```

```python
X_train = X_train.astype({col: 'int' for col in X_train.select_dtypes('bool').columns})
X_test = X_test.astype({col: 'int' for col in X_test.select_dtypes('bool').columns})
```

```python
import shap

explainer = shap.Explainer(best_xgb, X_train)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test)
```

```python
from xgboost import plot_importance
import matplotlib.pyplot as plt

# Plot feature importance
plt.figure(figsize=(10, 8))
plot_importance(best_xgb, max_num_features=20, importance_type='gain')
plt.title("Top 20 Feature Importances (by Gain)")
plt.xlabel('Average Gain')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Get predicted probabilities for the positive class
y_proba_tuned = best_xgb.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, _ = roc_curve(y_test, y_proba_tuned)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'XGBoost (AUC = {roc_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guess')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate (Recall)', fontsize=12)
plt.title('ROC Curve – Tuned XGBoost Model', fontsize=14)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

The ROC curve illustrates the trade-off between the true positive rate (recall) and the false positive rate for different thresholds. Our tuned XGBoost model achieves an outstanding AUC of 0.993, which confirms that it is highly capable of distinguishing between churned and non-churned customers. The sharp upward curve followed by a flat line shows that the model maintains high recall even at low false positive rates — an ideal scenario in churn prediction.



The analysis of customer churn reveals several actionable insights. 🔍 Key Findings:
Transaction behavior is the top driver of churn. Customers with fewer transaction counts (Total_Trans_Ct) and lower total amounts spent (Total_Trans_Amt) are more likely to churn.

Churn risk is higher among customers with lower relationship depth, such as fewer products (Total_Relationship_Count) or longer inactivity (Months_Inactive_12_mon).

Customer segments matter. Income groups like “Less than $40K” and marital status (especially single customers) show elevated churn rates.

The Platinum card group, while premium, had higher churn, possibly due to unmet expectations.
